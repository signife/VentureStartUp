```python
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
iris_data = iris.data
iris_df = pd.DataFrame(data = iris.data ,columns = iris.feature_names)
iris_df.head()
```

```python
# @title 표준화 ? 시키는 방법
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(iris_df) #학습

iris_scaled = scaler.transform(iris_df) #적용
iris_scaled
```

```python
# @title 정규화 시키는방법
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(iris_df)
iris_scaled = scaler.transform(iris_df)

df_scaled = pd.DataFrame(iris_scaled, columns = iris.feature_names)
df_scaled
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

dataset= load_iris()

data = dataset['data']
target = dataset['target']

#train_test_split
X_train, X_test, y_train, y_test = train_test_split(data,target,test_size = 0.2, shuffle=True, stratify = target, random_state = 34)
print(len(X_train))
print(len(X_test))
print(len(y_train))
print(len(y_test))
```
