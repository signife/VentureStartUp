```python
import pandas as pd
from math import log
from konlpy.tag import Okt
```

```python
okt = Okt()

data = '''
나는 영어를 좋아 합니다.
영어, 수학을 매일 공부합니다.
과학 공부를 좋아하고 수학도 좋아 합니다.
영어,수학,과학을 좋아 하고 영어는 매일 공부합니다.
'''
```

```python
words = okt.morphs(data)
words

result = []
for temp in words:
    if len(temp) > 1:
        result.append(temp)

result = list(set(result))
result
```

```python
documents = data.split('.')
documents
```

```python
d_list = []
for temp in documents:
    if temp.strip() != '':
        d_list.append(temp.strip())

d_list
```

```python
N = len(d_list)

#tf구하는 함수
def tf(t,d):
    return d.count(t)

#idf구하는 함수
def idf(t):
    df = 0
    for doc in d_list:
        df += t in doc

    return log(N/(df+1))

def tfidf(t,d):
    return tf(t,d) * idf(t)
```

```python
word_list = result

t_result = []
for temp in range(N):
    t_result.append([])
    d = d_list[temp]

    for temp1 in range(len(word_list)):
        t = word_list[temp1]
        t_result[-1].append(tfidf(t,d))

t_result
```

```python
df = pd.DataFrame(t_result,columns = word_list)
df
```

```python
tfidf_result = []
for temp in range(len(word_list)):
    t = word_list[temp]
    tfidf_result.append(idf(t))


df_idf = pd.DataFrame(tfidf_result, index=word_list,columns=['IDF'])
df_idf
```
