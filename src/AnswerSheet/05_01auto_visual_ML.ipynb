```python
av = AutoViz_Class()

av.AutoViz(
    filename='',
    dfte = df,
    depVar = 'target',
    verbose=1, #0 간단히, 1:자세히 2:파일로
    max_rows_analyzed=df.shape[0],
    max_cols_analyzed=df.shape[1]
)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=34,stratify=df.iloc[:,-1])
```

```python
print(len(X_train))
print(len(X_test))
```

```python
df.iloc[:,-1]
```

```python
from lazypredict.Supervised import LazyClassifier
from lazypredict.Supervised import LazyRegressor

models = LazyClassifier(verbose=0, predictions=True)
#mocdls = LazyRegressor(verbose=0, predictions=True)

model_result,predictions = models.fit(X_train,X_test,y_train,y_test)
```
