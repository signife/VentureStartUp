```python
import pandas as pd

df = pd.read_csv('appreply.csv')
df = df.iloc[3:]
df = df.reset_index()
df = df.drop(columns=['index','Unnamed: 0'])
df
```

```python
from konlpy.tag import Okt
import re
from tqdm import tqdm

okt = Okt()

def tokenize_text(text):
    text = re.sub(r'[^ㄱ-ㅣ가-힣\s]','',text)

    okt_morphs = okt.pos(text)

    words = []
    for word,pos in okt_morphs:
        if pos in ['Adjective','Verb','Noun']:
            words.append(word)

    word_str = ' '.join(words)
    return word_str

token_list = []

for temp in tqdm(df['text']):
    token_list.append(tokenize_text(temp))
```

```python
token_list
```

```python
corpus_list = []

for index in range(len(token_list)):
    corpus = token_list[index]
    if len(set(corpus.split())) < 3:
        corpus_list.append(corpus)

corpus_list

for corpus in corpus_list:
    token_list.remove(corpus)


print(len(token_list))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

count_vectorizer = CountVectorizer(max_df = 0.1, max_features = 1000, min_df=2, ngram_range=(1,2))

feat_vect = count_vectorizer.fit_transform(token_list)
feat_vect.shape
count_vectorizer.vocabulary_
#feat_vect.toarray()
```

```python
lda = LatentDirichletAllocation(n_components=5)
lda.fit(feat_vect)
```

```python
lda.components_[0][600]
```

```python
lda.components_[0].argsort()[::-1][0:10]
```

```python
lda.components_
```

```python
def display_topic(model, featur_name,num_top_words):
    for topic_index, topic in enumerate(model.components_):
        print('Topic ',topic_index+1)
        topic_word_indexes = topic.argsort()[::-1]
        top_index = topic_word_indexes[0:num_top_words]
        print(top_index)
        feature_concat = [featur_name[i] for i in top_index]
        print(feature_concat)

feature_name = count_vectorizer.get_feature_names_out()
display_topic(lda,feature_name,10)

```

```python
import pyLDAvis.lda_model

pyLDAvis.enable_notebook()
vis = pyLDAvis.lda_model.prepare(lda,feat_vect,count_vectorizer)
pyLDAvis.display(vis)
```

```python
sent_topic = lda.transform(feat_vect)
sent_topic
```

```python
sent_topic_per_list = []

for temp in range(sent_topic.shape[0]):
    sent_topic_list = sent_topic[temp].argmax()
    topic_per = sent_topic[temp].max()
    sent_topic_per_list.append([temp, sent_topic_list, topic_per])

topic_df = pd.DataFrame(sent_topic_per_list,columns = ['no','토픽번호','확률'])
doc_topic_df = topic_df.join(df)
doc_topic_df
```
