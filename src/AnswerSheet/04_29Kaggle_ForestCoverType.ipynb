```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/삼육대2024수업 자료/covtype.csv')
df
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df.iloc[:,-1] = le.fit_transform(df.iloc[:,-1])
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1],stratify=df.iloc[:,-1],random_state=34)
y_train
```

```python
df.iloc[:,-1].unique()
```

```python
model = XGBClassifier(random_state=34,device='cuda')
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import f1_score

pred = model.predict(X_test)
pred
```

```python
f1_score(y_test,pred,average='micro')
```
