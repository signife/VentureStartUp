```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = load_iris()
```

```python
df_iris = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df_iris
```

```python
kmeans = KMeans(n_clusters=3,max_iter=300,random_state=34)
kmeans.fit(df_iris)
```

```python
kmeans.labels_
```

```python
df_iris['cluster'] = kmeans.labels_
df_iris['target'] = data.target
df_iris
```

```python
iris_result = df_iris.groupby(['target','cluster'])['sepal length (cm)'].count()
```

```python
iris_result
```

```python
kmeans.cluster_centers_
```

```python
df_iris = pd.DataFrame(data.data,columns=data.feature_names)
df_iris
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(data.data)
pca_result
```

```python
df_iris['pca_x'] = pca_result[:,0]
df_iris['pca_y'] = pca_result[:,1]
df_iris
```

```python
df_iris['target'] = data.target
df_iris
```

```python
df_iris['cluster'] = kmeans.labels_
df_iris
```

```python
cluster_0 = df_iris[df_iris['cluster']==0].index
cluster_1 = df_iris[df_iris['cluster']==1].index
cluster_2 = df_iris[df_iris['cluster']==2].index
```

```python
plt.scatter(x=df_iris.loc[cluster_0,'pca_x'] , y=df_iris.loc[cluster_0,'pca_y'] , marker='o')
plt.scatter(x=df_iris.loc[cluster_1,'pca_x'] , y=df_iris.loc[cluster_1,'pca_y'] , marker='s')
plt.scatter(x=df_iris.loc[cluster_2,'pca_x'] , y=df_iris.loc[cluster_2,'pca_y'] , marker='^')
plt.show()
```
