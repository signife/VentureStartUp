```python
import pandas as pd
df = pd.read_excel('seoul.xls')
df.columns = ['학교명','위도','경도']
df
```

```python
seoul_map = folium.Map(location = [37.5662,126.9785],zoom_start=12,tiles = 'OpenStreetMap')

for temp in range(len(df['학교명'])):
    name = df.iloc[temp]['학교명']
    lat = df.iloc[temp]['위도']
    lng = df.iloc[temp]['경도']

    folium.Marker(location = [lat,lng],popup=name).add_to(seoul_map)

seoul_map.save('서울지역대학교현황.html')

seoul_map
```
