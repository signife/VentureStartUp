```python
import numpy as np
import pandas as pd
from sklearn import datasets

data = datasets.load_diabetes()
data
```

```python
data.keys()
```

```python
df = pd.DataFrame(data['data'],columns = data['feature_names'])
df['target'] = data['target']
df
```

```python
df.corr()
```

```python
import seaborn as sns

sns.heatmap(df.corr())
plt.show()
```

```python
import matplotlib.pyplot as plt

X = df.bmi.values
Y = df.target.values

plt.scatter(X,Y, alpha = 0.5)
plt.xlabel('비만')
plt.ylabel('정답데이터')
plt.show()
```

```python
import scipy.stats as stats
X = df.bmi.values
Y = df.target.values

stats.pearsonr(X,Y)
#피벨류 0.5이하 > 귀무가설 상관관계 없다가 틀림.
```
