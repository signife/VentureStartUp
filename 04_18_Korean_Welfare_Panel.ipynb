```python
job_name = pd.read_excel('/content/drive/MyDrive/sdata/(2021년 16차 한국복지패널조사) 조사설계서-가구용(beta3).xlsx',sheet_name='직종코드(2019 신분류)')
job_name.rename(columns={'Unnamed: 3': '직업'},inplace = True)
job_name = job_name[['소분류','직업']]
job_name.rename(columns = {'소분류':'job_code', '직업': 'job'}, inplace = True)
job_df = data.merge(job_name,how='left',on ='job_code')
job_df
```

```python
job_income_mean = job_df.dropna(subset = ['job','income']).groupby('job',as_index = False).agg(mean_income = ('income','mean'))
```

```python
top_income_20 = job_income_mean.sort_values('mean_income',ascending= False).head(20)
sns.barplot(data = top_income_20, x = 'job', y = 'mean_income')
plt.xticks(rotation=90)
plt.show()
```
