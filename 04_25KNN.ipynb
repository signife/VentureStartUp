```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

data = load_breast_cancer()
X_train,X_test,y_train,y_test = train_test_split(data.data, data.target, stratify = data.target, random_state = 34)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
print(knn.score(X_train, y_train))
print(knn.score(X_test,y_test))
```

```python

from sklearn.model_selection import GridSearchCV

knn_model = KNeighborsClassifier()
parameters = {'n_neighbors':range(1,31)}
knn_grid = GridSearchCV(knn_model, param_grid = parameters, cv = 5)

knn_grid



```

```python
knn_grid.fit(X_train, y_train)
```

```python
knn_grid.best_estimator_
```

```python
result = pd.DataFrame(knn_grid.cv_results_['params'])
result['mean_test_score'] = knn_grid.cv_results_['mean_test_score']
result
```

```python
result.sort_values('mean_test_score',ascending = False)
```
